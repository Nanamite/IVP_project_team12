```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
image = Image.open('archive/Set12/01.png')
image = np.array(image)
plt.imshow(image,cmap='gray')
```

```python
def spatial_noise(image):
    sigmas = [0,15,25,50]
    row, col = image.shape
    noise = np.zeros(image.shape)
    for i in range(4):
        noise[int(i/2)*int(row/2):int(i/2+1)*int(row/2), (i%2)*int(col/2):(i%2+1)*int(col/2)] = np.random.normal(0, sigmas[i], (int(row/2), int(col/2)))
    noisy_image = image + noise
    return noisy_image
```

```python
noisy_image = spatial_noise(image)
```

```python
plt.imshow(noisy_image,cmap='gray')
```

```python
Path('archive/Set12_noisy').mkdir(exist_ok=True)
for img in Path('archive/Set12').iterdir():
    image = Image.open(img)
    image = np.array(image)
    noisy_image = spatial_noise(image)
    plt.imsave('archive/Set12_noisy/'+img.name, noisy_image, cmap='gray')
```
